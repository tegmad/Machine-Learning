### Data preprocessing 

```python
df_final = pd.read_csv('krisha.csv')
df_final
```

```python
df_final.info()
```

```python
df_final.duplicated().sum()
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final = df_final[['Краткая информация', 'Тип дома','Город', 'Жилой Комплекс', 'Год Постройки', 'Площадь м2', 'Безопасность', 'Этаж', 'Цена']]
df_final
```

```python
df_final['Комната'] = df_final['Краткая информация'].str[0]
df_final
```

```python
df_final['Район'] = df_final['Город'].str.split(', ').str[1]
df_final
```

```python
df_final['Площадь м2'] = df_final['Краткая информация'].str.extract(r'(\d+\.?\d*)\s*м²')
df_final.head()
```

```python
df_final['Цена в 〒'] = df_final['Цена'].str.replace('\n〒','')
df_final = df_final.drop(columns=['Цена'])
df_final.head()
```

```python
df_final.dropna(subset=['Этаж'], inplace=True)
```

```python
df_final[['Этаж_кв', 'Этаж_макс']] = df_final['Этаж'].str.split(' из ', expand=True)
```

```python
df_final = df_final.drop('Этаж', axis=1)
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
df_final = df_final.drop('Жилой Комплекс', axis=1)
df_final = df_final.drop('Безопасность', axis=1)
df_final.dropna(subset=['Тип дома'],inplace=True)
df_final.dropna(subset=['Район'],inplace=True)
df_final = df_final.dropna(subset=['Этаж_макс'])
```

```python
df_final.info()
```

```python
df_final['Цена в 〒'] = df_final['Цена в 〒'].str.replace(' ', '').str.replace(',', '').astype(int)
```

```python
df_final['Тип дома'] = df_final['Тип дома'].astype(str)
df_final['Город'] = df_final['Город'].astype(str)
df_final['Район'] = df_final['Район'].astype(str)
df_final['Комната'] = df_final['Комната'].astype(int)
df_final['Год Постройки'] = df_final['Год Постройки'].astype(int)
df_final['Площадь м2'] = df_final['Площадь м2'].astype(float)
df_final['Этаж_кв'] = df_final['Этаж_кв'].astype(float).astype(int)
df_final['Этаж_макс'] = df_final['Этаж_макс'].astype(int)
df_final.info()
```

```python
df_final['Город'] = df_final['Город'].str.split(',').str[0]
```

```python
df_final.to_csv('krisha_cleaned.csv', index=False)
```

```python
import numpy as np

def filter_by_iqr(df, column):
    Q1 = np.percentile(df[column], 25)
    Q3 = np.percentile(df[column], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df_astana_filtered = filter_by_iqr(df_astana, 'Цена в тг')
df_almaty_filtered = filter_by_iqr(df_almaty, 'Цена в тг')
df_shymkent_filtered = filter_by_iqr(df_shymkent, 'Цена в тг')

df_final_filtered = pd.concat([df_astana_filtered, df_almaty_filtered, df_shymkent_filtered])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df_final_filtered['Цена в тг (млн)'] = df_final_filtered['Цена в тг'] / 1_000_000

def plot_city_histogram(city, color, ax, max_price):
    # Фильтруем данные для города
    city_data = df_final_filtered[df_final_filtered['Город'] == city]['Цена в тг (млн)']
    
    mean_val = city_data.mean()
    median_val = city_data.median()
    mode_val = city_data.mode()[0] if not city_data.mode().empty else None
    
    sns.histplot(city_data, kde=True, color=color, ax=ax, label='Histogram')
    
    ax.axvline(mean_val, color='green', linestyle='--', label=f'Mean: {mean_val:.2f} Million Tenge')
    ax.axvline(median_val, color='red', linestyle='--', label=f'Median: {median_val:.2f} Million Tenge')
    
    if mode_val is not None:
        ax.axvline(mode_val, color='blue', linestyle='--', label=f'Mode: {mode_val:.2f} Million Tenge')
    
    ax.set_title(f'Distribution Price In  {city}', fontsize=14)
    ax.set_xlabel('Price In Million Tenge', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.set_xlim(0, max_price)  
    ax.legend(fontsize=10)
    ax.grid(True)

cities = ['Астана', 'Алматы', 'Шымкент']
colors = sns.color_palette("husl", len(cities))
max_price = df_final_filtered['Цена в тг (млн)'].max()  

fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

for i, (city, color) in enumerate(zip(cities, colors)):
    plot_city_histogram(city, color, axes[i], max_price)

plt.tight_layout()
plt.suptitle('Distibution Prices In Million Tenge', fontsize=16, y=1.02)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

X = df_almaty_encoded.drop('Цена в тг',axis=1)
y = df_almaty_encoded['Цена в тг']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

rf = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],  
    'max_depth': [10, 20, 30, None], 
    'min_samples_split': [2, 5, 10],  
    'min_samples_leaf': [1, 2, 4],  
    'bootstrap': [True, False]  
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='neg_mean_absolute_error')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print(f"Best Parameters: {best_params}")

best_rf_model = grid_search.best_estimator_

y_pred = best_rf_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Random Forest - Mean Absolute Error (MAE): {mae:.2f}")
print(f"Random Forest - R^2 Score: {r2:.2f}")

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='green', alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Real Prices In Million Tenge')
plt.ylabel('Predicted Prices In Million Tenge')
plt.title('Real vs Predicted Prices (Random Forest)')
plt.show()
```
